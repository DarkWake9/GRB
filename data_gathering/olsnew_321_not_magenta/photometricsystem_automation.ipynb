```python
###### import file with gcn and corresponding system info generated by Sam's crawling code #### 
df_newsys = pd.read_csv('GRB_GCNs_system_info.txt',sep='\t',usecols=[0,1,2])
df_newsys['System'] = df_newsys['System'].astype('string')
df_newsys['GCN'] = df_newsys['GCN'].astype('string')
df_newsys = df_newsys.drop_duplicates(subset=['GCN'],keep='first').reset_index(drop=True)
# boolean = df_newsys['GCN'].duplicated().any() #double check for duplicates
pd.set_option('display.max_rows', None)
```

# Old format files

```python
location = 'C:/Users/biagi/Documents/automatizeGRB/test_for_Navonil/'
df_list = []# create an empty list that would contain dataframes
filenames = glob2.glob(location+'*mag*.txt')

for file in filenames:
    print(file)
    df_list.append(pd.read_csv(file,sep='\s+', names=['Time(s)','Mag','MagErr','Filter','Telescope','GalExtCorr','Source']))
    #sep='\s+'
    #'Time(s)','Mag','MagErr','Filter','Telescope','GalExtCorr','Source'
    #"Time","Mag",'MagErr','Filter','Source'
    
    
for i in range(len(df_list)):
    swift_list = ['white_swift','u_swift','v_swift','uvm2_swift','b_swift','uvw1_swift','uvw2_swift']
    df_list[i].insert(loc=4, column='System', value='none') #since no system column exists previously
    df_list[i]['System'] = df_list[i]['System'].astype('string')
    df_list[i]['Source'] = df_list[i]['Source'].astype('string')
    
    vals =  set(df_newsys['GCN']).intersection(df_list[i]['Source']) #finds intersecting gcn
    for j in vals: # Replacing routine from known system info list
        old = df_newsys.loc[df_newsys['GCN'] == j,'System']
        df_list[i].System = np.where(df_list[i]['Source'] == j, old.values[0], 'none')
    df_list[i].loc[df_list[i]['Filter'].isin(swift_list),'System'] = 'VEGA' #swift in VEGA system
    df_list[i] = df_list[i].sort_values(by=['Time(s)'], ascending=True).reset_index(drop=True)
    globals()[f"df{i}"] = df_list[i]

names = [os.path.basename(x) for x in glob.glob(location+'*mag*.txt')]
#print(names[2])
out_path = 'C:/Users/biagi/Documents/automatizeGRB/test_for_Navonil/outputs/'
for i in range(len(df_list)):
    df_list[i].to_csv(out_path+names[i], sep='\t',index=False)
```

```python
location = '/home/angana/GRBLC/automatic_system_update/179_ONLY_APPROVED_magfiles/'
df_list = []# create an empty list that would contain dataframes
filenames = glob2.glob(location+'*mag*.txt')

for file in filenames:
    df_list.append(pd.read_csv(file,sep='\s+', names=["Time","Mag",'MagErr','Band/Filter','System','Telescope/Instrument','GalExtCorr','Source']))
for i in range(len(df_list)):
    df_list[i]['System'] = df_list[i]['System'].astype('string')
    df_list[i]['Source'] = df_list[i]['Source'].astype('string')
```

```python
#### For NO existing system info in magnitude file, it extracts information from system file ###

for i in range(len(df_list)):  #list of dataframes (mag files)
    df_list[i].loc[df_list[i]['Telescope/Instrument'] == 'Swift/UVOT', 'System'] = 'VEGA'
    a = df_list[i].loc[df_list[i]['System'] == 'none','Source']
    vals =  set(df_newsys['GCN']).intersection(a)
    for j in vals:
        replacement = df_newsys.loc[df_newsys['GCN'] == j,'System'].values        
        old = df_list[i].loc[df_list[i]['Source'] == j,'System'].values
        df_list[i].replace(old,replacement[0], inplace=True)
```

```python
############# Sorts the magnitude file according to time and save it to text file ###########

for i in range(len(df_list)): #splits list of pandas into separate pandas
    df_list[i] = df_list[i].sort_values(by=['Time'], ascending=True).reset_index(drop=True)
#     df_list[i].to_csv('file_name.csv',index=False) #MUST SAVE IT TO TEXT FILE WITH APPROPRIATE NAME
    globals()[f"df{i}"] = df_list[i]
    
names = [os.path.basename(x) for x in glob.glob(location+'*mag*.txt')]
out_path = '/home/angana/GRBLC/automatic_system_update/outputs/'
for i in range(len(df_list)):
    df_list[i].to_csv(out_path+names[i], sep='\t',index=False)
```
